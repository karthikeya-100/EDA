```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading the dataset
apps=pd.read_csv("../googleplaystore.csv")

# printing the head of the dataset

apps.head()
```

```python
# There are 10841 rows and 13 columns in the dataset
apps.shape
```

***Description of the Column Names in the dataset***<br>
***App***-Name of the App<br>
***Category***-Which category that particular app belongs to<br>
***Rating***-Rating given to that app on scae 1-5<br>
***Size***-Size require to install the App<br>
***Installs***- How many downlods of a app so far<br>
***Type***-Type of the app whether it is paid or Free<br>
***Content Rating***- Rating of the content in the App, Specified whether App can be used by everyone or only to a particular audience<br>
***Genre***-Specified to which genre App belongs to<br>
***Last Updated***- last updation date of the App<br>
***Current Ver***-Current version of the App<br>
***Android Ver***-Minimum Android version needs to download the App<br>

```python
# Checking whether dataset has duplicate values

apps.duplicated().any()
```

```python
# removing duplicate values in the dataset and resetting the index

apps_final=apps.drop_duplicates().reset_index()
```

```python
apps_final.head()
```

```python
# Here "index" column is created after resetting the index of dataframe. Because for the dataframe already default index exists
#making index column as default index for the dataframe

apps_final.set_index("index",inplace=True)
```

```python
apps_final.head()
```

```python
apps_final.info()
```

```python
# Finding if there are any missing values in the dataset

apps_final.isna().sum()
```

```python
apps_final.dtypes
```

From the above information, There are few missing values in the ***Rating,Type,Content Rating,Current Ver and Android Ver***<br> columns which needs to treat at first place. Also, Data types of columns ***Reviews,Installs, Price, Last Updated*** are misclassified which we need to convert them into proper datatypes

```python
# Let's work on treating missing values in the "Type" Column first

# Getting the entry in which Type column is NaN

apps_final[apps_final["Type"].isna()]
```

As we can see here, ***Price*** column of this particular record in "zero" which shows that this app is of Type ***Free***.<br> Replacing NaN in Type column with string "Free"

```python
apps_final.loc[(apps_final["Type"].isna()),"Type"]="Free"
```

```python
apps_final.isna().sum()
```

Removed NaN values from Type Column

```python
# Let's work on Content Rating Column 

# getting entries which have Content Rating column as NaN

apps_final[apps_final["Content Rating"].isna()]
```

As we can see, This particular record contains wrong data in each column, example Category=1.9 which doesn't exists. So we can drop or remove this whole row

```python
apps_final.drop(10472,inplace=True)
```

```python
apps_final.isna().sum()
```

Removed NaN value from Content Rating column

```python
# Let's work on Current Ver column

# getting rows in which Current Ver column is NaN

apps_final[apps_final["Current Ver"].isna()]
```

There 8 missing values in the Current Ver column. As we cannot make any assumptions on the "Current Ver" of any App, It's better to replace NaN values with string "Varies with Device"

```python
apps_final.loc[(apps_final["Current Ver"].isna()),"Current Ver"]="Varies with device"
```

```python
apps_final.isna().sum()
```

Removed NaN values from Current Ver column

```python
# Let's work on the Android Ver column

# Getting entries where Android Ver column is NaN

apps_final[apps_final["Android Ver"].isna()]
```

As same as Current Ver column, We cannot make any assumptions for the minimum android version to install the App, Better to replace NaN values with string "Varies with device"

```python
apps_final.loc[(apps_final["Android Ver"].isna()),"Android Ver"]="Varies with device"
```

```python
apps_final.isna().sum()
```

Removed null values from the Android Ver Column

```python
# Let's now work on NaN values in the Rating Column

# Getting entries which have null values in Rating Column

apps_final[apps_final["Rating"].isna()]
```

There are lot of missing values in the Rating column, It's good to replace those NaN values with the "Average Rating" of the Category to which that App belongs to

```python
# storing entries in which Rating column is Null

null_rating=apps_final[apps_final["Rating"].isna()]
```

```python
null_rating.head()
```

```python
# Getting unique Categories of the Null rating entries

null_rating["Category"].unique()
```

```python
# Getting mean rating of each app category

mean_category_rating=apps_final.groupby("Category").mean()
```

```python
mean_category_rating
```

```python
mean_category_rating.loc[null_rating["Category"]]["Rating"].values.tolist()
```

```python
# replacing NaN values with average rating of the category to which app belongs to

null_rating["Rating"]=mean_category_rating.loc[null_rating["Category"]]["Rating"].values
```

```python
null_rating.head()
```

```python
null_rating.isna().sum()
```

```python
# Assigning Null rating back to main dataframe

apps_final[apps_final["Rating"].isna()]=null_rating
```

```python
apps_final.isna().sum()
```

Removed NaN values from Rating column

```python
apps_final.head()
```

```python
apps_final.dtypes
```

```python
# Converting reviews Column to int64 dataType
apps_final["Reviews"]=apps_final["Reviews"].astype("int")
```

```python
apps_final.dtypes
```

```python
apps_final.head()
```

If we observe Install column, every value is in the format of "Number+". For example "10,000+" means there are more than Ten Thousand downloads for that App. Which means there are minimum of ten thousand downloads for that App. So its better to create a new Column "Minimum Downloads" which contain minimum downloads of a particular App

```python
apps_final.loc[:,"Minumum Dowloads"]=apps_final["Installs"]
```

```python
apps_final.head()
```

```python
# Removing Special characters in the Minimum downloads and converting into numeric data type

chars_to_remove=["+",","]
for char in chars_to_remove:
    apps_final["Minumum Dowloads"]=apps_final["Minumum Dowloads"].astype("str").str.replace(char,"")
```

```python
# Converting into numeric data type
apps_final["Minumum Dowloads"]=pd.to_numeric(apps_final["Minumum Dowloads"])
```

```python
apps_final.head()
```

```python
apps_final.dtypes
```

```python
# Dropping the Installs Column from the dataset
apps_final.drop("Installs",axis=1,inplace=True)
```

```python
apps_final.head()
```

```python
# Removing special character $ in price column and converting column into numeric

apps_final["Price"]=apps_final["Price"].astype("str").str.replace("$","")
apps_final["Price"]=pd.to_numeric(apps_final["Price"])
```

```python
apps_final.head()
```

```python
apps_final.dtypes
```

```python
# Converting Last updated column to datetime object
apps_final["Last Updated"]=pd.to_datetime(apps_final["Last Updated"])
```

```python
apps_final.dtypes
```

```python
apps_final.head()
```

As we can see in the android version column, Entries are in the format of "version and up" or "version-version". For example "4.0.3 and up" means, To install that particular app android version should be 4.0.3 or above. So it's better to create a new column "Minimum_Android_version" which contains the version of android require to install that App


```python
# Assigning android version column to new column

apps_final.loc[:,"Minimum_Android_Version_req"]=apps_final["Android Ver"]
```

```python
pd.Series(["4.0.3 and up","Varies with device","4.0.3 - 7.8"]).apply(lambda x:x if x=="Varies with device" else str(x).split()[0])
```

```python
apps_final["Minimum_Android_Version_req"]=apps_final["Minimum_Android_Version_req"].apply(lambda x:x if x=="Varies with device" else str(x).split()[0])
```

```python
apps_final.head(10)
```

```python
# we can drop off the Android Ver column

apps_final.drop("Android Ver",axis=1,inplace=True)
```

# Exploratory Data Analysis

```python
apps_cleaned=pd.read_csv("GooglePlayStoreCleaned.csv")
apps_cleaned.head()
```

```python
apps_cleaned.dtypes
```

```python
apps_cleaned.isna().sum()
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot,init_notebook_mode
init_notebook_mode(connected=True)
```

```python
# Exploring the categories which dominates the App market

# We will be using plotly library for the visualization

#printing total number of unique categories
 
unique_categories=len(apps_cleaned["Category"].unique())
print(f"Number of unique categories of Apps  are {unique_categories}")
print("And they are Below")
print(apps_cleaned["Category"].unique().tolist(),end="\n")

#Counting number of apps in each category

num_of_apps=apps_cleaned["Category"].value_counts().sort_values(ascending=False)

#plotting which categories are dominating the App market

trace0=go.Bar(
    x=num_of_apps.index,
    y=num_of_apps.values
)

layout=go.Layout(
    title="Categories Vs Number of Apps",
    xaxis=dict(title="Categories of Apps"),
    yaxis=dict(title="Number of Apps")
)

data=[trace0]

fig=go.Figure(data,layout)

plotly.offline.iplot(fig)
```

As we can see above, There are very high number of Family category apps followed by Game Category<br>
And there are less number of Apps in the Beauty Category

```python
# Finding most reviewed app category


# Getting maximum reviews of each app category
num_of_reviews_per_cat=apps_cleaned.groupby("Category")["Reviews"].max().values


trace1=go.Bar(
    x=num_of_apps.index,
    y=num_of_reviews_per_cat
)
layout1=go.Layout(
    title="Max Number of Reviews Vs Category",
    xaxis=dict(title="Category of Apps"),
    yaxis=dict(title="Max Number of Reviews")
)
data1=[trace1]

fig1=go.Figure(data1,layout1)

plotly.offline.iplot(fig1)
```

As we can see that, House and Home category apps are most reviwed by the users 

```python
# Which category has highest number of ratings

number_of_ratings_per_cat=apps_cleaned.groupby("Category")["Rating"].mean().values


trace2=go.Bar(
    x=num_of_apps.index,
    y=number_of_ratings_per_cat,
)

# Setting the default value for the color

trace2.marker.color="grey"


#Finding index where highest number of average rating occurs

index=trace2["y"].tolist().index(max(number_of_ratings_per_cat))


# Changing color of the bar at that position

colors=[trace2.marker.color]*len(num_of_apps.index)
colors[index]="purple"


layout2=go.Layout(
    title="Average Rating Vs Category of Apps",
    xaxis=dict(title="Category of Apps"),
    yaxis=dict(title="Average rating")
)

data2=[trace2]

# Assigning colors list to trace color 

trace2.marker.color=colors


fig2=go.Figure(data2,layout2)

plotly.offline.iplot(fig2)
```

We can see that Sports Category app has maximum average rating

```python
# Let's see average rating of the Apps

average_rating=apps_cleaned["Rating"].mean()
print("Average rating of the Apps is ", average_rating)


trace3=go.Histogram(
    x=apps_cleaned["Rating"]
)

layout3=go.Layout(
    title="Distribution of rating",
    xaxis=dict(title="Frequency"),
    yaxis=dict(title="Ratings")
)
#Setting number of bins

trace3.nbinsx=70 


data3=[trace3]

fig3=go.Figure(data3,layout3)

plotly.offline.iplot(fig3)
```
